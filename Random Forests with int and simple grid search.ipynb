```python
import os
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
```

```python
print(test.describe())
train = train.drop(['Name', 'Cabin'], axis=1)
test = test.drop(['Name', 'Cabin'], axis=1)

```

```python
train['Age'] = np.nan_to_num(train['Age'])
test['Age'] = np.nan_to_num(test['Age'])
test['Fare'] = np.nan_to_num(test['Fare'])

for df in [train, test]:
    df['Sex_binary']=df['Sex'].map({'male':1, 'female':0})
    df['Sex x Age'] = df['Sex_binary'] * df['Age']
    df['Sex x Fare'] = df['Sex_binary'] * df['Fare']
    df['Fare x Age'] = df['Fare'] * df['Age']

print(df.describe())
    


features = ['Pclass', 'Age', 'Sex_binary', 'SibSp', 'Parch', 'Fare', 'Sex x Age', 'Sex x Fare', 'Fare x Age']
target = 'Survived'
```

```python
tuned_parameters = [{'n_estimators': [5, 10, 100, 500], 'max_features': ['auto', 5, 10, 50]}]

clf = GridSearchCV(
    RandomForestClassifier(), tuned_parameters
)
```

```python
clf.fit(train[features], train[target])
print(clf.best_params_)
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r"
          % (mean, std * 1.96, params))
        
```

```python
print(train.info())
```

```python
predictions = clf.predict(test[features])
```

```python
submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': predictions})
```

```python
filename = 'Titanic Predictions 3.csv'
```

```python
submission.to_csv(filename, index=False)
```
